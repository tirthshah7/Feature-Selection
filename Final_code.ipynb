```python
import os
import sys
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier 
from sklearn import metrics
from sklearn.model_selection import StratifiedKFold,cross_val_score
```

```python
scaler = StandardScaler()

dataset = list(map(int,input('Enter datset file number: ').split()))
classifier = input('Enter classifier name (knn/rf/mlp): ')
if classifier =='knn':
    clf = KNeighborsClassifier(n_neighbors = 8, metric = 'minkowski', p = 2)
elif classifier == 'rf':
    clf = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
elif classifier == 'mlp':
    clf = MLPClassifier(hidden_layer_sizes=(10,10,10), max_iter=1000)
else:
    print('enter appropriate option')
    sys.exit()
fs_technique = input('Enter Feature Selection Method (ilfs,innfs,ecfs): ')
n_features = int(input('Enter the number of features to be selected: '))
```

```python
for i in dataset:
    filename = os.path.join('D:\TIRTH\Windsor Study Material\SEM-3\AML\Project\Datasets','D'+str(i)+'.csv')
    df = pd.read_csv(filename, header=None)

    X = df[df.columns[:-1]]
    Y = df[df.columns[-1]] 
    if fs_technique == 'ilfs':
        ilfs = pd.read_excel(r'D:\TIRTH\Windsor Study Material\SEM-3\AML\Project\ranks.xlsx', sheet_name='ILFS')
        sel = list(ilfs.iloc[i-1])[1:df.shape[1]-1]
        if 354 in sel:
            sel.remove(354)
        elif 221 in sel:
            sel.remove(221)
        new_X = X.iloc[:,sel[:n_features]]
        
    elif fs_technique == 'innfs':
        inffs = pd.read_excel(r'D:\TIRTH\Windsor Study Material\SEM-3\AML\Project\ranks.xlsx', sheet_name='INFFS')
        sel = list(inffs.iloc[i-1])[1:df.shape[1]-1]
        if 354 in sel:
            sel.remove(354)
        elif 221 in sel:
            sel.remove(221)
        new_X = X.iloc[:,sel[:n_features]]
        
    elif fs_technique == 'ecfs':
        ecfs = pd.read_excel(r'D:\TIRTH\Windsor Study Material\SEM-3\AML\Project\ranks.xlsx', sheet_name='ECFS')
        sel = list(ecfs.iloc[i-1])[1:df.shape[1]-1]
        if 354 in sel:
            sel.remove(354)
        elif 221 in sel:
            sel.remove(221)
        new_X = X.iloc[:,sel[:n_features]]

    skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=1)
    j=1
    for train_idx, test_idx in skf.split(new_X, Y):
        print('fold',j)
        x_train, x_test = np.array(new_X.iloc[train_idx]),np.array(new_X.iloc[test_idx])
        y_train, y_test = np.array(Y.iloc[train_idx]),np.array(Y.iloc[test_idx])
        if classifier == ' MLP':
            scaler.fit(x_train)
            x_train = scaler.transform(x_train)  
            x_test = scaler.transform(x_test)
        clf.fit(x_train, y_train)
        y_pred = clf.predict(x_test)
        acc = clf.score(x_test, y_test)
        f1 = f1_score(y_test, y_pred, average='weighted')
        print('Accuracy :'.format(j), acc*100)
        print('F-Mearure:'.format(j),f1)
        print()
        j+=1
    print(new_X)
```
